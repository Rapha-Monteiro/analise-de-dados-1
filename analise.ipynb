```python
import pandas as pd 


# Criar um data frama(trazer os dadaos para a memória do python)

df = pd.read_csv('dados.csv')

display(df)
```

```python
#head imprime apenas uma amostra dos dados
print(df.head(2))

print()

#info
print(df.info())
      
print('\n')
    
print(df.describe())

print('\n')

print(df.shape)

print(df.columns)
```

```python
# Consultar apneas uma coluna
marca = df['marca']

print(marca)

print('\n')

#consulta direto no print
print(df['modelo'])

#Consultar mais que uma coluna 
marca_modelo_cor = df[['marca','modelo','cor']]

display(marca_modelo_cor)

```

```python
# Criar filtros

# Quero saber quantos carros não pagam mais ipva no meu estoque

noIpva = df[df['ano'] < 2006 ]

display(noIpva)
```

```python
# Apresente os carros que custam mais que R$20.000

Carrosmaiscaros = df[df['valor'] > 20000]

display(Carrosmaiscaros)

# Apresente apenas os carros da chevrolet

carrosGM = df[df['marca'] == 'Chevrolet']

display(carrosGM)
```
